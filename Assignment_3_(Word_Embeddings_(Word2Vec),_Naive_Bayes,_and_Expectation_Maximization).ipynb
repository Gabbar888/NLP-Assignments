## **1. Introduction to Word2Vec**

### **What is Word2Vec?**
Word2Vec is a shallow neural network that learns **dense vector representations of words** such that semantically similar words lie close together in the embedding space.

The key idea: instead of representing words as one-hot vectors, we represent them as low-dimensional **embeddings** learned by predicting context words from a target word (or vice versa).

---
### Skip-gram Architecture

The **Skip-gram model** in Word2Vec learns word embeddings by predicting context words given a center (target) word. Instead of using one-hot vectors, we use Label Encoding, which is more memory-efficient and computationally optimal.

####  Intuition

Sentence:

If target word = `"cat"`, context = `["The", "sat"]`  
→ Training pairs:
- (`cat` → `The`)

#### Architecture and Training Flow (with Index-based Input)

Let:  
- \( V \): size of the vocabulary  
- \( d \): embedding dimension  
-  t in $ 0, 1, \dots, V-1 $: index of the target word  
-  $E \in {R}^{V \times d}$: input embedding matrix  
-  $E' \in {R}^{V \times d}$: output embedding matrix

The integer index \( t \) is used to perform a direct embedding lookup in matrix \( E \), avoiding one-hot vector multiplication.





---

####  Step-by-Step Computation:

1. **Input**: Integer ID of target word \( t \)

2. **Embedding Lookup** (instead of one-hot):

$$
u = E[t] \in \mathbb{R}^d
$$

3. **Score for Context Word \( c \)**:

$$
\text{score}(c) = u^\top E'[c]
$$

4. **Softmax over Vocabulary**:

$$
P(w_c \mid w_t) = \frac{\exp(u^\top E'[c])}{\sum_{w=1}^{V} \exp(u^\top E'[w])}
$$

---



#### Objective Function

Given a **center word** $ w_t $ , the Skip-gram model aims to **predict surrounding context words** within a fixed window size.


$$
{L}_{\text{skip-gram}} = \sum_{-k \le j \le k,\ j \ne 0} \log P(w_{t+j} \mid w_t)
$$


where


| Term &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;| Description |
|------|-------------|
| $ {L}_{{skip-gram}} $ | The **log-likelihood** objective for predicting the context words surrounding the center word $ w_t $ This is the quantity we aim to **maximize** during training. |
| $ k $| The **window size**, which defines how many words before and after the center word are considered as context. |
| $ j $ | The **offset** from the center word $ t $, ranging from $ -k $ to $ +k $, excluding $ j = 0 $ (which would be the center word itself). |
| $ w_t $ | The **center word** (target word) whose surrounding words we are trying to predict. |
| $ w_{t+j} $ | A **context word** at position $ t+j $ relative to the center word $ w_t $ |
| $ P(w_{t+j} \mid w_t) $ | The **conditional probability** that word $ w_{t+j} $ appears in the context of $ w_t $. This is typically modeled using a softmax function or approximated using negative sampling. |
| $ \log P(w_{t+j} \mid w_t) $ | The **log-probability** of correctly predicting a context word. Using the log helps with numerical stability and converts the product of probabilities into a sum. |



---




#### Parameters Learned

- $ E \in {R}^{V \times d} $: Embedding matrix (input)
- $ E' \in {R}^{V \times d} $: Output matrix (context prediction)

**Embedding Lookup** via integer index replaces costly one-hot vector multiplication.

**Note**: Final embeddings are typically taken from matrix $ E $
---


#### Summary

- **Input**: Integer index of target word  
- **Output**: Predict context word indices  
- **Lookup**: Directly fetch embedding vector from matrix  
- **Train**: Using softmax or negative sampling  
- **Goal**: Words appearing in similar contexts should have similar vectors

### CBOW Architecture

The **CBOW model** in Word2Vec learns word embeddings by predicting a center (target) word given its surrounding context words. Like Skip-gram, it avoids one-hot vectors using index-based embedding lookup, making it both space- and time-efficient.

---

#### Intuition

Sentence:  
**"The cat sat on the mat"**, with **window size = 2**

If target word = `"cat"`, context = `["The", "sat"]`  
→ Training pairs:
- (`["The", "sat"]` → `cat`)

#### Architecture and Training Flow (with Index-based Input)

Let:  
- $ V $: size of the vocabulary  
- $ d $: embedding dimension  
-   C = $ c_1, c_2, \dots, c_m $ : list of context word indices  
-  $ t $: index of the target word  
-  $ E \in {R}^{V \times d} $ : input embedding matrix  
-  $ E' \in {R}^{V \times d} $ : output embedding matrix

Each context word $ c_i \in C $ is used to lookup a vector from $ E $, which are then averaged to form a single context embedding.

---

#### Step-by-Step Computation:

1. **Input**: Integer IDs of context words  C = $ {c_1, c_2, \dots, c_m} $

2. **Embedding Lookup**:

$$
h = \frac{1}{m} \sum_{i=1}^{m} E[c_i] \in \mathbb{R}^d
$$

3. **Score for Target Word $ t $**:

$$
\text{score}(t) = h^\top E'[t]
$$

4. **Softmax over Vocabulary**:

$$
P(w_t \mid \text{context}) = \frac{\exp(h^\top E'[t])}{\sum_{w=1}^{V} \exp(h^\top E'[w])}
$$

---

#### Objective Function

Given **context words** $ \{w_{t-k}, \dots, w_{t-1}, w_{t+1}, \dots, w_{t+k}\} $, the CBOW model aims to **predict the center word** \( w_t \).

$$
{L}_{\text{CBOW}} = \log P(w_t \mid w_{t-k}, \dots, w_{t-1}, w_{t+1}, \dots, w_{t+k})
$$

where


| Term &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    | Description |
|---------|-------------|
| $ {L}_{\text{CBOW}} $ | The **log-likelihood** objective for predicting the center word $ w_t $ given its surrounding context words. |
| $ k $ | The **window size**, which defines how many words before and after the target word are used as context. |
| $ c_i $ | The **context word index** used for embedding lookup. |
| $ w_t $ | The **center word** (target word) to be predicted. |
| $ P(w_t \mid \text{context}) $ | The **conditional probability** of $ w_t $ being the correct word, given the context. |
| $ log P(w_t \mid \text{context}) $ | The **log-probability** of predicting $ w_t $ correctly given its context. |


---

#### Parameters Learned

- $ E \in {R}^{V \times d} $: Embedding matrix (input)
- $ E' \in {R}^{V \times d} $: Output matrix (used for prediction)

**Note**: Final embeddings are typically taken from matrix $ E $

---

### Summary

- **Input**: Integer indices of context words  
- **Output**: Predict index of target (center) word  
- **Lookup**: Embed each context word, then average  
- **Train**: With softmax or negative sampling  
- **Goal**: Words that appear in similar contexts should learn similar vector representations

### **Negative Sampling in Word2Vec**

---

#### Why Not Full Softmax?

In the original Skip-gram formulation, the softmax function is used to compute the probability of a context word given a target word:

$$
P(w_o \mid w_t) = \frac{\exp(u_{w_o}^\top v_{w_t})}{\sum_{w=1}^{V} \exp(u_w^\top v_{w_t})}
$$

Where:
- $ v_{w_t} $: input (center) embedding of word $ w_t $
- $ u_{w_o} $: output (context) embedding of word $ w_o $
- $ V $: vocabulary size

**Problem**:  
The denominator sums over **all words in the vocabulary** $ V $ — this is extremely slow for large corpora.

---

#### Solution: Negative Sampling

Instead of updating **all output weights**, **update only a small number of "negative samples"** along with the one positive pair.

---

#### Intuition

- For each training pair $ (w_t, w_o) $:
  - Treat it as a **positive example**
  - Sample $ k $ random words $ \{w_1^{'}, w_2^{'}, ..., w_k^{'}\} $ from the vocabulary — these are **negative samples**
- The goal:
  - **Maximize** the probability of real context word $ w_o $
  - **Minimize** the probability of negative samples $ w_i^{'} $

---

#### Loss Function

For a center word $ w_t $ and a context word $ w_o $, the **negative sampling loss** is:

$$
{L}_{\text{negative-sampling}} = -\log \sigma(u_{w_o}^\top v_{w_t}) - \sum_{i=1}^{k} \log \sigma(-u_{w_i^{'}}^\top v_{w_t})
$$

Where:
- $ \sigma(x) = \frac{1}{1 + \exp(-x)} $  : sigmoid function
- $ u_w $: output embedding  of word $ w $
- $ v_w $: input embedding  of word $ w $
- $ w_o $: actual word (positive sample)
- $ w_i^{'} $: i-th negative sample
- $ k $: number of negative samples

---

#### Training Workflow (Step-by-Step)

1. **Input**: target word $ w_t $, context word $ w_o $
2. **Get embeddings**:
   - $ v_{w_t} \in {R}^d $ from input matrix $ E $
   - $ u_{w_o} \in {R}^d $ from output matrix $ E'$
3. **Sample** $ k $ random words from vocabulary as negative samples
4. **Compute loss** using the formula above
5. **Backpropagate** and update only:
   - $ v_{w_t} $
   - $ u_{w_o} $
   - $ u_{w_i^{'}} $ for all $ i \in [1, k] $

---




#### Why It Works

- Drastically reduces **computation time**.
- Still allows the model to learn **good semantic embeddings**.
- Scales to **billions of tokens** and **millions of words**.
- Enables **mini-batch training** with sparse updates.


---
### **Hierarchical Softmax in Word2Vec**

---

####  Why Another Alternative to Softmax?

Like Negative Sampling, **Hierarchical Softmax (HS)** addresses the computational inefficiency of the full softmax, especially for large vocabularies.

**Key Idea**:  
 Instead of scoring all words in the vocabulary, we build a **binary tree** and compute the probability of a word by traversing the path from the **root to that word's leaf**.

---

#### How Hierarchical Softmax Works

1. Build a **binary tree** where:
   - **Leaves** represent vocabulary words.
   - **Internal nodes** make **binary decisions**: left/right.

2. Each word is uniquely identified by a **path from the root to a leaf**.

3. The model:
   - Learns vector representations for each **internal node**.
   - Predicts the correct **binary decision** at each node in the path.

---

#### Example

Say the word **"sat"** is located along this binary path:
```
Root → Left → Right → Left → [sat]
```

We must predict:
- Step 1: Go **Left**
- Step 2: Go **Right**
- Step 3: Go **Left**

So the model predicts **a sequence of decisions** — **not** the word directly.

---

#### Mathematical Formulation

Let:
- $ w $ : target word to predict
- $ w_t $ : input word
- $ L(w) $ : length of the binary path to $ w $
- $ n_i $ : the $ i^{th} $ internal node along the path to $ w $
- $ s_i \in \{+1, -1\} $: direction at node $ n_i $ (+1 = left, -1 = right)
- $ v_{w_t} $ : input vector for $ w_t $
- $ v_{n_i} $ : vector associated with internal node $ n_i $
- $ \sigma(x) = \frac{1}{1 + \exp(-x)} $

Then:

$$
P(w \mid w_t) = \prod_{i=1}^{L(w)} \sigma\left( s_i \cdot v_{n_i}^\top v_{w_t} \right)
$$

---

#### Loss Function

For predicting word $ w $ given center word $ w_t $, the **loss is**:

$$
{L}_{\text{HS}} = - \sum_{i=1}^{L(w)} \log \sigma\left( s_i \cdot v_{n_i}^\top v_{w_t} \right)
$$

- Similar to Negative Sampling, the model uses **sigmoid** activations at each node.
- Each decision is treated as a **binary classification task**.

---


#### Summary

- Hierarchical Softmax is a tree-based approximation to full softmax.
- Words are predicted by following binary paths.
- Each step in the path is trained via sigmoid + binary cross-entropy.
- Enables fast, full-vocab prediction with logarithmic complexity.

````python

## Preprocessing and Vocabulary
class Vocabulary:
    #Used for handling vocabulary creation, frequency counting, and prep data for Word2Vec
    def __init__(self, min_count=5):
        self.min_count = min_count
        self.stoi = {}
        self.itos = []
        self.word_freqs = Counter()
        self.unigram_table = []
        self.vocab_size = 0

    def build_vocab(self, corpus):
        # We create a List of all tokens a word_frequency table
        # and prepare their indexes and mapping
        print("building vocabulary...")
        all_tokens = [token for sentence in corpus for token in sentence]
        self.word_freqs = Counter(all_tokens)
        sorted_words = sorted(self.word_freqs.items(), key=lambda x: x[1], reverse=True)
        self.itos = [word for word, freq in sorted_words if freq >= self.min_count]
        self.stoi = {word: i for i, word in enumerate(self.itos)}
        self.vocab_size = len(self.itos)
        print("Completed building vocabulary")
        print(f"Vocabulary size: {self.vocab_size} words")

    def prepare_negative_sampling(self, table_size=1e8, power=0.75):
        # We create a probabilities unigram table to do negative sampling
        print("Creating unigram table for negative sampling...")
        freqs = np.array([self.word_freqs[word] for word in self.itos])
        powered_freqs = freqs ** power
        probabilities = powered_freqs / np.sum(powered_freqs)
        self.unigram_table = np.random.choice(len(self.itos), size=int(table_size), p=probabilities)
        print("Unigram table created")

    def numericalize(self, corpus):
        # We convert the whole corpus as a list of sentence
        # which is list of indices of the token
        return [[self.stoi[token] for token in sentence if token in self.stoi] for sentence in corpus]

````

````python
## Huffman Tree For Hierarchial Softmax
class HuffmanNode:
    def __init__(self, word_idx=None, freq=None, left=None, right=None, parent=None):
        self.word_idx = word_idx # Index according to vocab.itos (only for leaf nodes)
        self.freq = freq
        self.left = left
        self.right = right
        self.parent = parent

    # For heapq comparison
    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(vocab):
  # Use to build a huffman tree from the vocab frequencies
  # Returns path which is a list of parent node indices
  # and code which is a list of binary decisions: 0->right, 1->left

    print("Started Building Huffman Tree...")
    #leaf node for each word
    node_heap = [HuffmanNode(word_idx=i, freq=vocab.word_freqs[vocab.itos[i]])
                 for i in range(vocab.vocab_size)]
    heapq.heapify(node_heap)

    # Building the tree based on higher frequency assigned to lower length path
    # We use standard greedy algo for huffman tree building
    internal_node_idx_counter = 0
    while len(node_heap) > 1:
        node1 = heapq.heappop(node_heap)
        node2 = heapq.heappop(node_heap)

        # Create a new internal node
        new_freq = node1.freq + node2.freq
        internal_node = HuffmanNode(freq=new_freq, left=node1, right=node2)

        # Assign parent pointers
        node1.parent = internal_node
        node2.parent = internal_node

        # Assign a unique index to the internal node different from word indices
        # Internal node indices will be 0 to vocab_size - 2 i.e n-1 internal nodes as there are n leaves
        internal_node_index = internal_node_idx_counter
        internal_node.node_index = internal_node_index # Store index in node
        internal_node_idx_counter += 1


        heapq.heappush(node_heap, internal_node)

    # The root is the last node remaining in the heap
    root = node_heap[0]

    # We traverse the tree to get paths and codes for each leaf node
    paths = {}
    codes = {}

    def traverse(node, current_path, current_code):
        if node.word_idx is not None:
            paths[node.word_idx] = current_path[:] # Store path (list of internal node indices)
            codes[node.word_idx] = current_code[:] # Store code (list of binary decisions)
            return

        # Traverse left (code 1)
        if node.left:
             traverse(node.left, current_path + [node.node_index], current_code + [1])

        # Traverse right (code 0)
        if node.right:
             traverse(node.right, current_path + [node.node_index], current_code + [0])


    traverse(root, [], [])

    print("Huffman Tree built")
    return paths, codes
````

````python
# Word2Vec Model Implementation
class Word2Vec:
    def __init__(self, vocab, mode, embedding_dim=100, window_size=2, learning_rate=0.03, epochs=4, k_negative=5, batch_size=512):
        # defined all required values and initialised input and output weight matrix
        self.vocab = vocab
        self.mode = mode
        self.embedding_dim = embedding_dim
        self.window_size = window_size
        self.initial_lr = learning_rate
        self.epochs = epochs
        self.k_negative = k_negative # number of negative sample per positive sample
        self.batch_size = batch_size
        self.word_count_actual = 0

        self.W_in = np.random.uniform(-0.5/embedding_dim, 0.5/embedding_dim, (vocab.vocab_size, embedding_dim)).astype(np.float64)

        if 'hs' in self.mode:
            self.W_out = np.zeros((vocab.vocab_size - 1, embedding_dim), dtype=np.float64)
            self.paths, self.codes = build_huffman_tree(self.vocab)
        else:
            self.W_out = np.zeros((vocab.vocab_size, embedding_dim), dtype=np.float64)

    def _generate_training_data(self, corpus):
      # creates training pairs by pairing centre word and context words
        if self.mode.startswith('cbow'):
            return self._generate_cbow_pairs(corpus)
        return self._generate_skipgram_pairs(corpus)

    def _generate_skipgram_pairs(self, corpus):
      # it takes the numericalized corpus and
      # returns a list of pairs of (centre_word,context_word) for all context_words
        pairs = []
        for s in corpus:
            for i, c in enumerate(s):
                for j in range(max(0, i - self.window_size), min(len(s), i + self.window_size + 1)):
                    if i != j: pairs.append((c, s[j]))
        return pairs

    def _generate_cbow_pairs(self, corpus):
      # it takes the same corpus and
      # returns a list of pairs of ([context_words],centre_word)
        pairs = []
        for s in corpus:
            for i, t in enumerate(s):
                ctx = [s[j] for j in range(max(0, i-self.window_size), min(len(s), i+self.window_size+1)) if i!=j]
                if ctx: pairs.append((ctx, t))
        return pairs

    def _sigmoid(self, x):
      # for getting the value of how much is the given pair
      # close to being a pair in real context
        return 1.0 / (1.0 + np.exp(-np.clip(x, -25, 25)))

    def _get_negative_samples(self, size):
        return self.vocab.unigram_table[np.random.randint(0, len(self.vocab.unigram_table), size=size)]

    def train(self, corpus):
      numerical_corpus = self.vocab.numericalize(corpus)
      training_data = self._generate_training_data(numerical_corpus)
      total_pairs = len(training_data)
      total_updates = total_pairs * self.epochs

      print(f"Starting training for '{self.mode}' with {total_pairs:,} total pairs...")

      for epoch in range(self.epochs):
          np.random.shuffle(training_data)
          with tqdm(total=total_pairs, desc=f"Epoch {epoch + 1}/{self.epochs}", unit="pair") as pbar:
              for i in range(0, total_pairs, self.batch_size):
                  batch = training_data[i : i + self.batch_size]
                  if not batch: continue
                  current_progress = (epoch * total_pairs) + i
                  lr = self.initial_lr * (1 - current_progress / total_updates)
                  lr = max(lr, self.initial_lr * (0.01)/100)
                  if self.mode == 'skipgram_ns': self._train_skipgram_ns_batch(batch, lr)
                  elif self.mode == 'skipgram_hs': self._train_skipgram_hs_batch(batch, lr)
                  elif self.mode == 'cbow_ns': self._train_cbow_ns_batch(batch, lr)
                  elif self.mode == 'cbow_hs': self._train_cbow_hs_batch(batch, lr)
                  pbar.update(len(batch))


    def _train_skipgram_ns_batch(self, batch, lr):
        center_words, context_words = zip(*batch)
        center_vectors = self.W_in[list(center_words)]
        pos_targets = np.array(context_words)
        neg_targets = self._get_negative_samples(size=(len(batch), self.k_negative))
        all_targets = np.hstack([pos_targets[:, None], neg_targets])
        labels = np.zeros_like(all_targets, dtype=np.float64); labels[:, 0] = 1.0
        target_vectors = self.W_out[all_targets]
        # ---  Forward Pass ---
        dots = np.einsum('ij,ikj->ik', center_vectors, target_vectors)
        sigs = self._sigmoid(dots)
        # --- Backward Pass ---
        errors = sigs - labels
        grad_W_out = np.einsum('ik,ij->ikj', errors, center_vectors)
        grad_W_in = np.einsum('ik,ikj->ij', errors, target_vectors)
        # --- Gradient Update---
        self.W_in[list(center_words)] -= lr * grad_W_in
        for i in range(len(batch)):
            self.W_out[all_targets[i]] -= lr * grad_W_out[i]

    def _train_skipgram_hs_batch(self, batch, lr):
        for center_word, context_word in batch:
            center_vector = self.W_in[center_word]
            path, codes = self.paths[context_word], self.codes[context_word]
            if not path: continue
            node_vectors = self.W_out[path]
            # ---  Forward Pass ---
            dots = np.dot(node_vectors, center_vector)
            sigs = self._sigmoid(dots)
            # --- Backward Pass ---
            grads = sigs - (1 - np.array(codes))
            error = np.dot(grads, node_vectors)
            # --- Gradient Update---
            self.W_out[path] -= lr * np.outer(grads, center_vector)
            self.W_in[center_word] -= lr * error

    def _train_cbow_ns_batch(self, batch, lr):
        for context_indices, target_word in batch:
            h = np.mean(self.W_in[context_indices], axis=0)
            all_targets = np.concatenate(([target_word], self._get_negative_samples(size=self.k_negative)))
            labels = np.array([1] + [0] * self.k_negative)
            target_vectors = self.W_out[all_targets]
            # ---  Forward Pass ---
            dots = np.dot(target_vectors, h)
            sigs = self._sigmoid(dots)
            # --- Backward Pass ---
            errors = sigs - labels
            grad_h = np.dot(errors, target_vectors)
            # --- Gradient Update---
            self.W_out[all_targets] -= lr * np.outer(errors, h)
            self.W_in[context_indices] -= lr * grad_h / len(context_indices)

    def _train_cbow_hs_batch(self, batch, lr):
        for context_indices, target_word in batch:
            h = np.mean(self.W_in[context_indices], axis=0)
            path, codes = self.paths[target_word], self.codes[target_word]
            if not path: continue
            node_vectors = self.W_out[path]
            # ---  Forward Pass ---
            dots = np.dot(node_vectors, h)
            sigs = self._sigmoid(dots)
            # --- Backward Pass ---
            grads = sigs - (1 - np.array(codes))
            grad_h = np.dot(grads, node_vectors)
            # --- Gradient Update---
            self.W_out[path] -= lr * np.outer(grads, h)
            self.W_in[context_indices] -= lr * grad_h / len(context_indices)

    def save(self, filename):
        with open(filename, "wb") as f:
            pickle.dump({"embeddings": self.W_in, "vocab": self.vocab.itos, "stoi": self.vocab.stoi}, f)
        print(f"Model saved to {filename}")

    def solve_analogy(self, w1, w2, w3):
        stoi, itos, embeddings = self.vocab.stoi, self.vocab.itos, self.W_in
        if w1 not in stoi or w2 not in stoi or w3 not in stoi: return "<unk>"
        v1, v2, v3 = embeddings[stoi[w1]], embeddings[stoi[w2]], embeddings[stoi[w3]]
        target_vec = v1 + v2 - v3
        similarities = cosine_similarity(target_vec.reshape(1, -1), embeddings).flatten()
        exclude_indices = {stoi.get(w) for w in [w1, w2, w3] if w in stoi}
        similarities[list(exclude_indices)] = -np.inf
        return itos[np.argmax(similarities)]

    def get_similar_words(self, word, top_n=10):
      #using cosine similarity get similar words to given word
        stoi, itos, embeddings = self.vocab.stoi, self.vocab.itos, self.W_in
        if word not in stoi: return []
        word_vec = embeddings[stoi[word]]
        similarities = cosine_similarity(word_vec.reshape(1, -1), embeddings).flatten()
        similarities[stoi[word]] = -np.inf
        top_indices = np.argsort(similarities)[::-1][:top_n]
        return [(itos[i], similarities[i]) for i in top_indices]


````

````python
def generate_plots(model, model_name, num_seed_words=20, top_n_similar=5):
    print(f"\nGenerating interactive plots for {model_name}...")

    all_alpha_words = [w for w in model.vocab.itos if w.isalpha()]
    if len(all_alpha_words) > num_seed_words:
        sample_words = np.random.choice(all_alpha_words, num_seed_words, replace=False)
    else:
        sample_words = all_alpha_words

    plot_words_set = set(sample_words)
    for word in sample_words:
        try:
            similar_words = [w for w, s in model.get_similar_words(word, top_n=top_n_similar)]
            plot_words_set.update(similar_words)
        except Exception as e:
            print(f"Could not get similar words for '{word}': {e}")

    plot_words = list(plot_words_set)
    plot_embeddings = np.array([model.W_in[model.vocab.stoi[w]] for w in plot_words])

    print(f"Plotting {len(plot_words)} unique words...")

    #dimensionality reduction
    tsne_coords = TSNE(
        n_components=2,
        random_state=42,
        perplexity=min(30, len(plot_words) - 1)
    ).fit_transform(plot_embeddings)

    umap_coords = umap.UMAP(
        n_components=2,
        n_neighbors=min(15, len(plot_words) - 1),
        random_state=42
    ).fit_transform(plot_embeddings)

    for method, coords in [('t-SNE', tsne_coords), ('UMAP', umap_coords)]:
        df = pd.DataFrame({'word': plot_words, 'x': coords[:, 0], 'y': coords[:, 1]})

        # Create the plot with hover info instead of persistent text to get rid of a lot of annotation
        fig = px.scatter(
            df,
            x='x',
            y='y',
            hover_name='word',  # Show word on hover
            text='word',
            title=f'Word Embeddings Visualized with {method} | {model_name}'
        )

        fig.update_traces(
            mode='markers',  # Start with markers only, no text
            marker=dict(size=8, opacity=0.7, line=dict(width=1, color='DarkSlateGrey')),
            textfont_size=10
        )

        filename = f"{model_name}_{method.lower()}.html"
        fig.write_html(filename)
        print(f" Saved interactive plot to {filename}")
````

### **Q1. Skip-gram vs CBOW Trade-offs**
Given a highly imbalanced corpus where rare words appear very few times and common words dominate, which architecture (Skip-gram or CBOW) would you prefer and why?  
Justify your answer using the mathematical objectives and data characteristics.

---------------- **write your answer here:** ----------------

For a text collection with many rare words, Skip-gram is more effective than CBOW.

In a highly imbalanced corpus, I would prefer **Skip-gram** because its objective better captures rare words. The CBOW model maximizes  

$$
J_{CBOW} = - \sum_{t=1}^T \log p(w_t \mid w_{t-m}, \dots, w_{t+m}),m \ne 0
$$  

so updates to a word embedding are directly tied to its frequency as a target, meaning rare words receive very few gradient updates. Skip-gram, on the other hand, maximizes  

$$
J_{SG} = - \sum_{t=1}^T \sum_{-m \le j \le m, j \ne 0} \log p(w_{t+j} \mid w_t)
$$  

where each occurrence of a word produces $(2m)$ context predictions. Thus, the expected updates for a word scale with $(\text{freq}(w) \times 2m)$, giving rare words multiple gradient contributions per appearance. This makes Skip-gram more effective than CBOW for learning good embeddings in highly imbalanced data where common words dominate.

### **Q2. Impact of Negative Sampling Distribution**
If negative samples are drawn uniformly at random versus proportionally to word frequency, how would the embeddings differ?  
Which strategy do you think better captures semantic relationships, and why?

---------------- **write your answer here:** ----------------

If negative samples are drawn uniformly at random, rare words are sampled more often than they naturally occur, which can distort embeddings by over-penalizing uncommon words and pushing them too far from their true semantic neighborhoods. On the other hand, sampling negative words in proportion to their frequency matches the natural distribution of the corpus, so common words like "the" or "is" are used more often as negatives, helping the model focus on distinguishing meaningful co-occurrences.

This frequency-based strategy generally produces embeddings that better capture semantic relationships, since it avoids artificially inflating the influence of rare words.

### **Q5. Computational Efficiency**
Word2Vec with softmax requires computing a normalization term over the entire vocabulary, which becomes computationally expensive for large $V$.  
Alternative training objectives such as **Negative Sampling** and **Hierarchical Softmax** are used to improve scalability.

- Explain why **negative sampling** makes training feasible for very large vocabularies.
- How does **hierarchical softmax**  reduce the complexity?
- Compare the computational complexity of the following methods:

  - **Full Softmax**:
  - **Negative Sampling**:
  - **Hierarchical Softmax**:

  Use Big-O notation

---------------- **write your answer here:** ----------------

Training Word2Vec with a standard softmax function is incredibly slow for large vocabularies because, for every single training sample, it must calculate a score for all words in the entire dictionary just to normalize the probability. This is computationally wasteful. Negative Sampling cleverly simplifies this problem by changing the task. Instead of asking "which word is it?", it asks a much easier binary question: "Is this pair of words a real context pair or a fake one?". The model is trained on the one true pair (the positive sample) and just a handful of randomly selected "fake" pairs (the negative samples). Since we only perform calculations for this small, fixed number of samples instead of the entire vocabulary, training becomes vastly more efficient.



Hierarchical Softmax offers another smart solution by organizing the entire vocabulary into a binary tree structure, with words as the leaves. Instead of one massive, computationally expensive calculation across all words, the model just has to make a series of simple left/right decisions to navigate from the root of the tree to the correct word leaf. This reduces the problem from a single massive choice to a short sequence of easy choices. In terms of computational complexity, a Full Softmax is $O(V)$, where V is the vocabulary size. Negative Sampling is much faster at $O(k)$, where k is the small number of negative samples (typically 5-20). Hierarchical Softmax is also extremely efficient, with a complexity of $O(log_2(V))$, which corresponds to the height of the tree. This complexity here is with respect to one softmax calculation.

# **Task 2 : Naive Bayes**


## **1. Concepts**  
Naive Bayes is a **probabilistic machine learning algorithm** based on **Bayes’ theorem** with the simplifying assumption that all features are conditionally independent given the class. Despite its simplicity, it is widely used in domains such as spam detection, sentiment analysis, and text classification due to its efficiency and effectiveness on high-dimensional data.  

---

## **1.1 Motivation**  

- Many classification tasks involve **large feature spaces** (e.g., words in documents).  
- Complex models struggle with high dimensionality and require significant computation.  
- Naive Bayes offers a **lightweight and interpretable** solution that is both **fast to train and test**.  
- Works surprisingly well even when the independence assumption does not hold perfectly.  

**Example:** Predict whether an email is *spam* or *ham* using word occurrences.  

---

## **1.2 Core Ideas**  

## Naive Bayes Theory

- **Bayes’ Theorem**

$$
P(C \mid X) = \frac{P(X \mid C) \, P(C)}{P(X)}
$$

Where:  
- \($C$\): Class label  
- \($X$\): Feature vector (e.g., tokens in a document)  

---

### **1. Naive Independence Assumption**

$$
P(X \mid C) = \prod_{i=1}^n P(x_i \mid C)
$$


- \($X = (x_1, x_2, \dots, x_n)$\): the feature vector (all attributes/words in a document).  
- \($C$\): the class label (e.g., spam/ham, positive/negative, topic).  
- \($x_i$\): the \($i$\)-th feature of \($X$\) (e.g., presence/absence or count of word \($i$\)).  
- \($n$\): the number of features (e.g., vocabulary size).  
- \($P(X \mid C)$\): probability of observing the full feature vector \($X$\) given class \($C$\).  
- \($P(x_i \mid C)$\): probability of observing the $i$-th feature given class \($C$\).  

👉 The **naive** assumption is that features are conditionally independent given the class, so we can multiply their probabilities.

---

### **2. Classification Rule**

$$
\hat{C} = \arg\max_C \; P(C) \prod_{i=1}^n P(x_i \mid C)
$$

- \($\hat{C}$\): predicted class for the instance (the classifier’s output).  
- \($\arg\max_C$\): choose the class \($C$\) that maximizes the expression.  
- \($P(C)$\): prior probability of class \($C$\) (from class frequencies).  
- \($\prod_{i=1}^n P(x_i \mid C)$\): likelihood of observing \($X$\) under class \($C$\).  

👉 The classifier picks the class that makes the features most probable under Bayes’ rule.


- **Intuition**  
Naive Bayes simplifies classification by assuming features are conditionally independent given the class. Instead of estimating a complex joint probability distribution, it multiplies individual feature likelihoods, making it efficient and effective for high-dimensional data like text.

## **1.4 Implementation Details**  

## Naive Bayes Steps

1. **Training Phase**  
   - Compute **prior probabilities** $P(C)$ from class frequencies.  
   - Estimate **conditional probabilities** $P(x_i \mid C)$ from feature counts.  

---

2. **Smoothing**  
- Use **Laplace smoothing** to avoid zero probabilities:  

   $$
   P(x_i \mid C) = \frac{\text{count}(x_i, C) + 1}{\sum_j \text{count}(x_j, C) + |V|}
   $$  
where,
-  $|V|$ = vocabulary size.  
- \($x_i$\): the \($i$\)-th feature of \($X$\) (e.g., presence/absence or count of word \($i$\)).  

---

3. **Prediction Phase**  
- Compute posterior probabilities for each class.  
- Use logarithms to prevent underflow in large feature spaces.  
- Select the class with maximum posterior probability.

````python
class TextClassifierNB:
    def __init__(self, smoothing_factor=1.0):
        #Initializes the classifier with a smoothing factor
        self.smoothing_factor = smoothing_factor
        self.log_class_priors = {}
        self.token_counts_per_class = {}
        self.total_tokens_per_class = {}
        self.vocabulary = set()
        self.class_labels = []

    @staticmethod
    def _preprocess_text(document):
        document = document.lower()
        # Find sequences of alphanumeric characters and apostrophes
        return re.findall(r"[a-z0-9']+", document)

    def fit(self, documents, targets):
        num_docs = len(targets)
        assert len(documents) == num_docs, "Documents and targets must have the same length."

        class_counts = Counter(targets)
        self.class_labels = sorted(class_counts.keys())

        # Calculate log class priors directly
        self.log_class_priors = {cls: np.log(class_counts[cls] / num_docs) for cls in self.class_labels}

        self.token_counts_per_class = {cls: Counter() for cls in self.class_labels}
        self.total_tokens_per_class = {cls: 0 for cls in self.class_labels}

        print("Training Naive Bayes classifier...")
        for doc, target in tqdm(zip(documents, targets), total=num_docs, desc="Fitting model"):
            tokens = self._preprocess_text(doc)
            self.token_counts_per_class[target].update(tokens)
            self.total_tokens_per_class[target] += len(tokens)
            self.vocabulary.update(tokens)

        self.vocab_size = len(self.vocabulary)
        print(f"Training complete. Vocabulary size: {self.vocab_size} tokens.")

    def _calculate_log_likelihood(self, tokens, cls):
        #We calculate the log likelihood of a document belonging to a specific class.
        #This is the log P(document | class).
        log_likelihood_val = 0.0

        # Denominator for the conditional probability calculation with smoothing
        denominator = self.total_tokens_per_class[cls] + self.smoothing_factor * self.vocab_size

        class_token_counts = self.token_counts_per_class[cls]

        for token, count in Counter(tokens).items():
            # Numerator for the conditional probability with smoothing
            numerator = class_token_counts.get(token, 0) + self.smoothing_factor

            # Add the log probability multiplied by the token's frequency in the document
            log_likelihood_val += count * np.log(numerator / denominator)

        return log_likelihood_val

    def predict(self, documents):
        predictions = []
        print("Making predictions...")
        for doc in tqdm(documents, desc="Predicting"):
            tokens = self._preprocess_text(doc)

            posterior_scores = {}
            for cls in self.class_labels:
                # Main idea of Naive Bayes: log P(class) + log P(document | class)
                log_prior = self.log_class_priors[cls]
                log_likelihood = self._calculate_log_likelihood(tokens, cls)
                posterior_scores[cls] = log_prior + log_likelihood

            # The predicted class is the one with the highest log posterior score
            predicted_class = max(posterior_scores, key=posterior_scores.get)
            predictions.append(predicted_class)

        return predictions
````

````python
def compute_accuracy(ground_truth, predictions):
    #Calculates the accuracy of predictions
    correct_predictions = sum(gt == p for gt, p in zip(ground_truth, predictions))
    return correct_predictions / len(ground_truth) if len(ground_truth) > 0 else 0.0

def calculate_prf_metrics(ground_truth, predictions):
    #Calculates precision, recall, and F1-score
    unique_labels = sorted(set(ground_truth) | set(predictions))
    metrics_per_label = {}

    for label in unique_labels:
        tp = sum(1 for gt, p in zip(ground_truth, predictions) if gt == label and p == label)
        fp = sum(1 for gt, p in zip(ground_truth, predictions) if gt != label and p == label)
        fn = sum(1 for gt, p in zip(ground_truth, predictions) if gt == label and p != label)

        precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
        f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0

        metrics_per_label[label] = {'precision': precision, 'recall': recall, 'f1': f1}

    avg_precision = np.mean([metrics['precision'] for metrics in metrics_per_label.values()])
    avg_recall = np.mean([metrics['recall'] for metrics in metrics_per_label.values()])
    avg_f1 = np.mean([metrics['f1'] for metrics in metrics_per_label.values()])

    return {
        'precision_macro': float(avg_precision),
        'recall_macro': float(avg_recall),
        'f1_macro': float(avg_f1),
        'per_label': metrics_per_label
    }

def write_predictions_to_csv(predictions, filepath="nb_predictions.csv"):
    """Saves a list of predictions to a CSV file."""
    output_path = Path(filepath)
    with output_path.open('w', newline='') as file:
        csv_writer = csv.writer(file)
        for prediction in predictions:
            csv_writer.writerow([prediction])
    return output_path.resolve()

def write_metrics_to_json(metrics_dict, filepath="nb_results.txt"):
    """Saves a dictionary of metrics to a text file in JSON format."""
    output_path = Path(filepath)
    with output_path.open('w') as file:
        file.write("Naive Bayes evaluation results\n")
        file.write(json.dumps(metrics_dict, indent=2))
    return output_path.resolve()
````

# **Task 3 : Expectation Maximization**

## **1. Concepts**

Expectation Maximization (EM) is a **probabilistic optimization algorithm** used to estimate parameters of statistical models when data has **latent (hidden) variables**. It alternates between assigning probabilities to hidden variables (E-step) and maximizing the likelihood of the parameters (M-step). EM is widely applied in clustering (e.g., Gaussian Mixture Models), missing data problems, and probabilistic inference.  

---

## 1.1 Motivation

- Real-world datasets often contain **incomplete or hidden information** (e.g., cluster assignments are unknown).  
- Direct maximization of the likelihood function becomes intractable due to hidden variables.  
- EM provides an **iterative framework** to estimate parameters efficiently by breaking the problem into two simpler steps.  

Example:  
Clustering points into multiple Gaussian distributions when class labels are unknown.  

---

## 1.2 Core Ideas

### (A) Theoretical Explanation

- **Likelihood with hidden variables**:  

$$
P(X \mid \theta) = \sum_Z P(X, Z \mid \theta)
$$  

Where:  
- \( $X$ \): Observed data  
- \( $Z$ \): Latent (hidden) variables  
- \( $\theta$ \): Model parameters  

- **E-step (Expectation)**:  
  Estimate the posterior distribution of hidden variables given current parameters:  

$$
Q(Z) = P(Z \mid X, \theta^{(t)})
$$  

- **M-step (Maximization)**:  
  Update parameters by maximizing the expected log-likelihood:  

$$
\theta^{(t+1)} = \arg\max_\theta \, \mathbb{E}_{Q(Z)}[\log P(X, Z \mid \theta)]
$$  

- **Intuition**:  
  The E-step “fills in” missing/hidden data with probabilities, while the M-step re-estimates parameters as if the hidden data were observed. This alternation improves the likelihood iteratively.  

---

### (B) Example: Gaussian Mixture Model (GMM) Clustering  

Given data points without labels:  

- **E-step**: Compute the probability that each data point belongs to each Gaussian component.  
- **M-step**: Update means, covariances, and mixture weights using these probabilities.  
- Repeat until convergence.  

This way, EM allows clustering without prior labels.  

---

## 1.3 Variants of EM  

1. **Gaussian Mixture Models (GMM-EM)**  
   - Clustering with Gaussian distributions.  

2. **Hidden Markov Models (HMM-EM, i.e., Baum-Welch Algorithm)**  
   - Sequence data with hidden states.  

3. **Soft K-Means**  
   - EM with isotropic Gaussians and uniform variance assumption.  

---

## 1.4 Implementation Details  

1. **Initialization**  
   - Start with random guesses for parameters (e.g., means and covariances in GMM).  

2. **E-step**  
   - Calculate the probability distribution of hidden variables given observed data.  

3. **M-step**  
   - Update parameter estimates by maximizing the expected log-likelihood.  

4. **Convergence**  
   - Stop when parameters stabilize or the log-likelihood improvement falls below a threshold.   

---

## 1.5 Summary  

- Expectation Maximization is an **iterative optimization algorithm** for parameter estimation in models with hidden variables.  
- Alternates between **E-step (inferring hidden variables)** and **M-step (optimizing parameters)**.  
- Widely applied in **clustering, HMMs, and incomplete-data problems**.  
- Sensitive to initialization but remains a cornerstone in unsupervised learning.

````python
import re
import csv
import json
import numpy as np
from collections import Counter
from pathlib import Path
from tqdm.auto import tqdm
````

````python
def extract_tokens(document_text):
    #Tokenizes a string into a list of alphanumeric tokens
    safe_text = document_text or ""
    return re.findall(r"[a-z0-9']+", safe_text.lower())

def create_vocabulary(documents, min_doc_freq=2, max_vocab_size=None):
    #Builds a vocabulary from a corpus of documents
    doc_frequency = Counter()
    for doc in documents:
        #Use a set to count each word only once per document
        doc_frequency.update(set(extract_tokens(doc)))

    #Filter tokens by minimum document frequency & trim
    frequent_tokens = [(token, count) for token, count in doc_frequency.items() if count >= min_doc_freq]
    frequent_tokens.sort(key=lambda item: -item[1])
    if max_vocab_size:
        frequent_tokens = frequent_tokens[:max_vocab_size]

    # Create the mapping from token to a unique integer index
    token_to_id = {token: i for i, (token, count) in enumerate(frequent_tokens)}
    return token_to_id

def vectorize_documents_sparse(documents, vocabulary):
    #Converts documents into a sparse count vector format
    sparse_vectors = []
    for doc in documents:
        token_counts = Counter(token for token in extract_tokens(doc) if token in vocabulary)

        if not token_counts:
            indices = np.array([], dtype=np.int32)
            values = np.array([], dtype=np.int32)
        else:
            indices = np.array([vocabulary[token] for token in token_counts.keys()], dtype=np.int32)
            values = np.array(list(token_counts.values()), dtype=np.int32)

        sparse_vectors.append((indices, values))
    return sparse_vectors
````

````python

class SparseTopicModelEM:
    # class for defining Multinomial Mixture Model which will be trained with EM
    def __init__(self, num_topics=10, dirichlet_prior=0.5, max_iter=100, tolerance=1e-4, seed=None):
        #Initializes the model
        self.n_topics = num_topics
        self.beta = dirichlet_prior  # Dirichlet prior for word distributions
        self.max_iter = max_iter
        self.tol = tolerance
        self.random_state = seed if seed is not None else 0
        self.dtype = np.float32

        # Model parameters to be learned
        self.topic_priors_ = None        # pi
        self.topic_word_dist_ = None     # phi
        self.log_topic_word_dist_ = None # log(phi)
        self.convergence_log_ = []

    def _initialize_parameters(self, vocab_size):
        rng = np.random.RandomState(self.random_state)
        self.topic_priors_ = np.ones(self.n_topics, dtype=self.dtype) / self.n_topics

        # Initialize word distributions for each topic from a Dirichlet distribution
        self.topic_word_dist_ = rng.dirichlet([1.0] * vocab_size, size=self.n_topics).astype(self.dtype)
        self.topic_word_dist_ = np.maximum(self.topic_word_dist_, 1e-12)
        self.topic_word_dist_ /= self.topic_word_dist_.sum(axis=1, keepdims=True)
        self.log_topic_word_dist_ = np.log(self.topic_word_dist_)

    def _compute_doc_log_likelihoods(self, token_indices, token_values):
        if token_indices.size == 0:
            return np.zeros(self.n_topics, dtype=self.dtype)
        # Efficiently compute \sum [count(w) * log P(w|topic)] for each topic
        relevant_log_probs = self.log_topic_word_dist_[:, token_indices]
        return (relevant_log_probs * token_values).sum(axis=1)

    def fit(self, sparse_vectors):
        vocab_size = 0
        if not hasattr(self, 'log_topic_word_dist_') or self.log_topic_word_dist_ is None:
            for indices, _ in sparse_vectors:
                if indices.size > 0:
                    vocab_size = max(vocab_size, int(indices.max()) + 1)
            self._initialize_parameters(vocab_size)

        last_log_likelihood = -np.inf

        for iteration in tqdm(range(self.max_iter), desc="EM Iterations"):
            topic_responsibilities_sum = np.zeros(self.n_topics, dtype=self.dtype)
            current_log_likelihood = 0.0
            m_step_numerator = np.zeros_like(self.topic_word_dist_)

            # --- E-Step ---
            for doc_indices, doc_values in tqdm(sparse_vectors, desc=f"E-Step (Iter {iteration+1})", leave=False):
                log_likelihood_per_topic = self._compute_doc_log_likelihoods(doc_indices, doc_values)
                log_joint_prob = log_likelihood_per_topic + np.log(self.topic_priors_ + 1e-16)

                # Log-Sum-Exp trick for numerical stability
                max_log = np.max(log_joint_prob)
                log_marginal = max_log + np.log(np.sum(np.exp(log_joint_prob - max_log)) + 1e-16)

                responsibilities = np.exp(log_joint_prob - log_marginal).astype(self.dtype)

                topic_responsibilities_sum += responsibilities
                current_log_likelihood += float(log_marginal)

                if doc_indices.size > 0:
                    update_values = responsibilities * doc_values[:, np.newaxis]
                    np.add.at(m_step_numerator.T, doc_indices, update_values)

            self.convergence_log_.append(current_log_likelihood)
            print(f"[EM] Iteration {iteration+1:03d}/{self.max_iter} - Log-Likelihood: {current_log_likelihood:.6f}")

            # --- M-Step ---
            # Update topic priors (pi)
            self.topic_priors_ = topic_responsibilities_sum / len(sparse_vectors)

            # Update topic-word distributions (phi) with Dirichlet smoothing
            m_step_numerator += self.beta
            m_step_denominator = m_step_numerator.sum(axis=1, keepdims=True)
            self.topic_word_dist_ = m_step_numerator / (m_step_denominator + 1e-16)
            self.log_topic_word_dist_ = np.log(self.topic_word_dist_ + 1e-16)

            # Check for convergence
            if iteration > 0 and abs(self.convergence_log_[-1] - last_log_likelihood) < self.tol:
                print("Convergence reached. Stopping iterations.")
                break
            last_log_likelihood = self.convergence_log_[-1]

        return self

    def predict(self, sparse_vectors):
        assignments = np.empty(len(sparse_vectors), dtype=np.int32)
        for i, (indices, values) in enumerate(tqdm(sparse_vectors, desc="Predicting Topics")):
            log_likelihood_per_topic = self._compute_doc_log_likelihoods(indices, values)
            log_joint = log_likelihood_per_topic + np.log(self.topic_priors_ + 1e-16)
            assignments[i] = int(np.argmax(log_joint))
        return assignments
````

````python

from datasets import load_dataset
train_data = load_dataset("Exploration-Lab/CS779-Fall25", "Assignment-3-em", split="train", token=hf_token)
test_data = load_dataset("Exploration-Lab/CS779-Fall25", "Assignment-3-em", split="test", token=hf_token)
train_documents = [x["text"] for x in train_data]
test_documents = [x["text"] for x in test_data]
test_ground_truth = [x.get("category") for x in test_data] if "category" in test_data.column_names else None

vocabulary = create_vocabulary(train_documents, min_doc_freq=1, max_vocab_size=5000)
print(f"Vocabulary created with {len(vocabulary)} tokens.")

train_vectors = vectorize_documents_sparse(train_documents, vocabulary)
test_vectors = vectorize_documents_sparse(test_documents, vocabulary)

topic_model = SparseTopicModelEM(num_topics=10, dirichlet_prior=0.5, max_iter=50, tolerance=1e-5, seed=42)
topic_model.fit(train_vectors)

topic_assignments = topic_model.predict(test_vectors)

output_path = Path("em_predictions.csv")
with output_path.open("w", newline="", encoding="utf-8") as f:
    csv_writer = csv.writer(f)
    for assignment in topic_assignments:
        csv_writer.writerow([int(assignment)])
print(f"\nWrote predictions to {output_path.resolve()}")

print("\n--- Model Summary ---")
print(f"Final log-likelihood: {topic_model.convergence_log_[-1]}")
print(f"Iterations run: {len(topic_model.convergence_log_)}")
print(f"Sample assignments: {topic_assignments[:20]}")
````

Q.1 Why is Expectation Maximization considered an iterative optimization algorithm, and how does it differ from direct likelihood maximization?

---------------- **write your answer here:** ----------------

In my view,Expectation Maximization (EM) is considered an iterative optimization algorithm because it alternates between two steps: the Expectation (E) step, where it estimates the missing or hidden data given current parameter values, and the Maximization (M) step, where it updates the parameters to maximize the expected likelihood computed in the E-step. This process repeats until convergence. Unlike direct likelihood maximization, which tries to maximize the likelihood function in one step (often difficult or intractable when data is incomplete or has hidden variables), EM breaks the problem into manageable iterations, gradually improving parameter estimates while handling latent information effectively.

Q.2 What is the role of the E-step and M-step in EM, and why are they repeated until convergence?  

---------------- **write your answer here:** ----------------

The role of the E-step (Expectation) is to act as the "soft assignment" step; it uses the current model parameters to make an educated guess about the hidden variables. For instance, in a clustering context, this means calculating the probability that each data point belongs to each cluster. The role of the M-step (Maximization) is then to take these probabilistic assignments as if they were true and update the model's parameters to maximize the likelihood of the data. So, you would recalculate the center and shape of each cluster based on these new assignments. This two-step process is repeated because each cycle of E-M is guaranteed to improve or maintain the model's log-likelihood. We continue this iterative refinement making better assignments, then creating better parameters until the parameters no longer change significantly, which means the algorithm has converged to a stable solution.

Q.3 Why is EM sensitive to initialization, and how can this problem be mitigated in practice?  

---------------- **write your answer here:** ----------------

In my experience, EM's sensitivity to initialization stems from the fact that it's fundamentally a "hill-climbing" algorithm. The likelihood function it navigates often has multiple peaks—what we call local maxima. EM is only guaranteed to find the top of the hill it starts on, which may not be the highest hill overall (the global maximum). A poor starting point can therefore trap the algorithm in a suboptimal solution. To mitigate this in practice, the most common and effective strategy is to use multiple random restarts. This involves running the EM algorithm several times, each with a different random initialization, and then selecting the model that results in the highest final log-likelihood.

Q.4 What are some advantages and disadvantages of EM compared to other clustering/optimization algorithms (e.g., K-Means, gradient-based methods)?  

---------------- **write your answer here:** ----------------

EM has several advantages over algorithms like K-Means or gradient-based methods. It allows soft assignments, so points can belong to multiple clusters with probabilities, and it can model clusters with different shapes and sizes using probabilistic distributions, unlike K-Means which assumes spherical clusters. EM also handles hidden or missing data naturally. However, it has some disadvantages: it is sensitive to initialization, can be computationally slower on large datasets, and relies on a parametric assumption that may not always match the real data. Overall, EM is more flexible in modeling complex cluster structures but comes with higher computational cost and sensitivity.
